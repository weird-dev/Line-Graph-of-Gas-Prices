```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,5), dpi=120)

plt.title('Gas Title Over Time (in USD)')

      #ANOTHER METHOD
# countries = ['USA', 'Canada', 'Soth Korea', 'Australia']
# for country in gas:
#     if country in countries:
#         plt.plot(gas.Year, gas[country], marker='.')
plt.plot(gas.Year, gas.USA, '.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'm.-', label='Australia')
plt.plot(gas.Year, gas.France, 'c.-', label='France')

plt.xlabel('Year')
plt.ylabel('Price(USD)')

plt.xticks(gas.Year[::3])

plt.legend()

plt.savefig('gas_prices.png', dpi=300)
plt.show()
```
